# Problem 1: Add two randomly generated NumPy 3D-matrices(tensor) using loops
HINT: B[i,j,k] = A1[i,j,k] + A2[i,j,k]

```python
import numpy as np
```

```python
Matrix1 = np.random.rand(2, 2, 2)
print(Matrix1)
print(Matrix1.shape)
```

```python
Matrix2 = np.random.rand(2, 2, 2)
print(Matrix2)
print(Matrix2.shape)
```

```python
Result = Matrix1 + Matrix2
print(Result)
print(Result.shape)
```

```python
(r, c, b) = Result.shape
print(r, c, b)
Result1 = np.zeros((r, c, b))
print(Result1)
```

```python
for i in range(len(Matrix1)):
    for j in range(len(Matrix1)):
        for k in range(len(Matrix1)):
            Result1[i][j][k] = Matrix1[i][j][k] + Matrix2[i][j][k]

print(Result1)
```

# Problem 2: Find if a point is inside or outside a rectangle.
HINT: How do you represent a rectangle?

```python
x1 = int(input("Enter 1st co-ordinate of diagonal of rectangle:"))
y1 = int(input("Enter 2nd co-ordinate of diagonal of rectangle:"))
x3 = int(input("Enter 5th co-ordinate of diagonal of rectangle:"))
y3 = int(input("Enter 6th co-ordinate of diagonal of rectangle:"))
print(x1, y1, x3, y3)
```

```python
x = int(input("Enter X co-ordinate of Point:"))
y = int(input("Enter Y co-ordinate of a point:"))
print(x, y)
```

```python
if x < x1 and x > x3 and y < y1 and y > y3:
    print("Poitn is Inside the Rectangle")
else:
    print("Point is Outside the Rectangle")
```

# Problem 4: Achieve the Array transformer using numpy.concatenate

```python
A = np.random.rand(5,4);
print(A);

B= np.random.rand(5,2);
print(B)
```

```python
C = np.concatenate((A, B), axis = 1)
print(C)
```

```python
D = np.random.rand(5,4);
print(D);

E = np.random.rand(4,4);
print(E)
```

```python
F = np.concatenate((D, E), axis = 0)
print(F)
```

```python
G = np.random.rand(5,4,3)
print(G.shape);

H = np.random.rand(5,3,3)
print(H.shape)

I = np.concatenate((G, H), axis = 1)
print(I.shape)
```
