# Problem 2: Compute maxima and minima of x^2+sin(x)+2*x

# Problem 4: Implement the logic we just discussed using simple loops for factorial of big number

```python
def fact(n):
    
    a = 1
    
    for i in range(1, n+1):
        a *= i
    return a
```

```python
print(fact(100))
```

```python
# Complete the extraLongFactorials function below.
def extraLongFactorials(n):
    cache = {}
    def rec(n):
        if n in cache:
            return cache[n]
        if n == 1:
            return 1
        result = n * rec(n-1)
        cache[n] = result
        return result
    return rec(n)

if __name__ == '__main__':
    n = int(input())

    print(extraLongFactorials(n))
```

# Problem 6: Number of Steps to Reduce a Number to Zero

```python
class Solution:
    def numberOfSteps(self, num: int) -> int:
        
        input_number = int(num)
    
        if input_number < 0:
            return 'Negative number not allowed'
    
        if input_number == 0:
            return 0
    
        if input_number == 1:
            return 1
    
        counter = 0
    
        while input_number > 0:
            
            if input_number % 2 == 0:
                input_number = input_number / 2
                counter = counter + 1

        
            elif input_number % 2 == 1:
                input_number == input_number / 2
                input_number = input_number - 1
                counter = counter + 1
            
            if input_number == 0:
                break
          
        return counter
    

    print(numberOfSteps(14,14))
```
